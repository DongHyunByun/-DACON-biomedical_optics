```python
import pandas as pd
import numpy as np
```

```python
csv=pd.read_csv("test.csv")
csv
```

```python
size=len(csv)
x_col=csv.columns[1:]
isNan=csv.isna()


x=csv[[i for i in x_col]].as_matrix()
x=[]
for i in range(size):
    temp=[]
    for col in x_col:
        #결측치면 평균으로
        if isNan[col][i]:
            temp.append(csv.mean()[col])
        else:
            temp.append(csv[col][i])
    x.append(temp)
x=np.array(x,dtype=np.float64)
```

```python
#모델 불러오기
from tensorflow.python.keras.models import load_model


model=load_model("1.softmax_model.h5")
```

```python
y=model.predict_proba(x)
```
